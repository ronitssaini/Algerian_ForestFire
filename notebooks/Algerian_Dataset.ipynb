🇨🇦Changing the required datatypes to int or whatever

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
# now to change datatpes like this by typing them manually is very tedious so we are gonna use loops for changing rest of datatypes from
# object to whatever(float in our case)

objects_of_interest=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects_of_interest
```

```python
for obj in objects_of_interest:
    if obj!='Classes':
        df[obj]=df[obj].astype(float)
```
